```python

import os
from dotenv import load_dotenv
```

```python
from dotenv import find_dotenv
load_dotenv(find_dotenv())
```

```python
AgentName = "Coding-Agent"
AgentInstructions = "You are a helpful assistant that can write and execute Python code to solve problems."
```

```python
client = FoundryChatClient(credential=AzureCliCredential())
code_interpreter_tool = client.get_code_interpreter_tool()
```

```python
agent = Agent(
    client=client,
    instructions=AgentInstructions,
    name=AgentName,
    tools=[code_interpreter_tool],
)

message = "Write a python function that returns a random even number between 1 and 100, and then call the function."

first_result = await agent.run(message)

print(f"Agent: {first_result.text}")
```
