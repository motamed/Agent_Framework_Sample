## 🔧 MCP Tool Integration Patterns

The reference sample demonstrates **two MCP tool integration patterns**:

### 🏃 **Pattern 1: Run-Level MCP Tools**
- MCP tools are defined and connected **before** running the agent
- Tools are passed to the `agent.run()` method directly
- Best for dynamic tool selection per query

### 🏗️ **Pattern 2: Agent-Level MCP Tools**
- MCP tools are passed when **creating** the agent
- The agent connects to MCP servers through its context manager (`async with agent`)
- Best for agents with fixed tool configurations

**Key Components:**
- 🔐 **AzureCliCredential**: Secure Azure authentication
- 🔌 **MCPStreamableHTTPTool**: HTTP-based MCP tool for external services
- 🏢 **FoundryChatClient**: Azure AI Foundry chat client

### Pattern 1: Pass MCP tools to the agent `run` method for dynamic tool usage

````python
# 🏃 Pattern 1: MCP Tools Defined on Run Level
# Tools are provided when running the agent — connect to MCP server first, then pass tools to run()

async with MCPStreamableHTTPTool(
    name="Microsoft Learn MCP",
    url="https://learn.microsoft.com/api/mcp",
) as mcp_server:
    agent = Agent(
        client=FoundryChatClient(credential=AzureCliCredential()),
        name="DocsAgent",
        instructions="You are a helpful assistant that can help with microsoft documentation questions.",
    )

    # First query
    query1 = "How to create an Azure storage account using az cli?"
    print(f"User: {query1}")
    result1 = await agent.run(query1, tools=mcp_server)
    print(f"{agent.name}: {result1}\n")

    print("\n=======================================\n")

    # Second query
    query2 = "What is Microsoft Agent Framework?"
    print(f"User: {query2}")
    result2 = await agent.run(query2, tools=mcp_server)
    print(f"{agent.name}: {result2}\n")
````

### Pattern 2: Pass MCP tools when creating the agent for fixed tool configurations

````python
# 🏗️ Pattern 2: MCP Tools Defined on Agent Level
# Tools are provided when creating the agent — the agent connects to MCP servers through its context manager

agent = Agent(
    client=FoundryChatClient(credential=AzureCliCredential()),
    name="DocsAgent",
    instructions="You are a helpful assistant that can help with microsoft documentation questions.",
    tools=MCPStreamableHTTPTool(
        name="Microsoft Learn MCP",
        url="https://learn.microsoft.com/api/mcp",
    ),
)

# Use agent as context manager to connect MCP tools
async with agent:
    # First query
    query1 = "How to create an Azure storage account using az cli?"
    print(f"User: {query1}")
    result1 = await agent.run(query1)
    print(f"{agent.name}: {result1}\n")

    print("\n=======================================\n")

    # Second query
    query2 = "What is Microsoft Agent Framework?"
    print(f"User: {query2}")
    result2 = await agent.run(query2)
    print(f"{agent.name}: {result2}\n")
````
